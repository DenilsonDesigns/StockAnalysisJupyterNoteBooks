### ETS decomposition (Seasonal Trends)

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
airline = pd.read_csv('../csv/airline_passengers.csv', index_col='Month')
```

```python
airline.head()
```

```python
airline.plot()
```

```python
airline.dropna(inplace=True)
```

```python
airline.index = pd.to_datetime(airline.index)
```

```python
airline.head()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(airline['Thousands of Passengers'], model='multiplicative')
```

```python
result.seasonal.plot()
```

```python
result.trend.plot()
```

```python
fig = result.plot()
```
